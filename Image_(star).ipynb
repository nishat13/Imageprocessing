```python
from matplotlib import pyplot as plt 
from skimage import data 
from skimage.feature import blob_dog, blob_log, blob_doh 
from math import sqrt
from skimage.color import rgb2gray
import glob 
from skimage.io import imread 
```

```python
image_name='star.jpg'
file_name="/content/drive/MyDrive/image processing/"+ image_name
```

```python
im = imread(file_name, as_gray=True)
plt.imshow(im, cmap=plt.cm.gray)
plt.show()
```

```python
blobs_log=blob_log(im, max_sigma=30, num_sigma=10,threshold=0.1)
blobs_log[:,2]=blobs_log[:,2]*sqrt(2)
num_rows=len(blobs_log)
print("total number of stars:", num_rows)
```
